Part d. Racial Makeup of the Giant Component (2 point)
-------------------

Using `sp.bincount()` compute the number of students of each race in the whole school and in the giant component.

```python
#Adding RaceN and GenderN attributes to giant component
for node in giantComponent.nodes(data=True):
    node[1]['RaceN'] = race_dict.get(node[1]['Race'])
    node[1]['GenderN'] = gender_dict.get(node[1]['Sex'])
GCRaceN = dict(nx.get_node_attributes(giantComponent, 'RaceN'))
print("Whole School:")
for race in race_dict:
    print('{}:\t {}'.format(race, sp.bincount(list(RaceN.values()))[race_dict.get(race)]))
print("\nGiant Component:")
for race in race_dict:
    print('{}:\t {}'.format(race, sp.bincount(list(GCRaceN.values()))[race_dict.get(race)]))
#print(sp.bincount(list(RaceN.values())))
#print(sp.bincount(list(GCRaceN.values())))
```

Compute the fraction of students of each race in the whole school and in the giant component.

```python
wholeTotal = len(fmh.nodes)
GCTotal = len(giantComponent.nodes)
print("Whole School:")


for race in race_dict:
    group = sp.bincount(list(RaceN.values()))[race_dict.get(race)]
    print('{}:\t {}'.format(race, group/wholeTotal))
print("\nGiant Component")
for race in race_dict:
    group = sp.bincount(list(GCRaceN.values()))[race_dict.get(race)]
    print('{}:\t {}'.format(race, group/GCTotal))
```

Plot the fractions of each race in the different components in a side-by-side bar plot. See Example Plot 1 in assignment handout.

```python
WholeFrac = [item/wholeTotal for item in sp.bincount(list(RaceN.values()))]
GCFrac = [item/GCTotal for item in sp.bincount(list(GCRaceN.values()))]

y = WholeFrac
z = GCFrac
width = 0.4
#Full Network
plt.bar(sp.arange(6)-width/2 -.05,y,width,align = 'center', color = 'red', label = "Full Network")
#Giant component
plt.bar(sp.arange(6)+width/2 +.05,z, width,align = 'center', color = 'blue', label = "Giant Component")
plt.legend()
plt.ylabel("% of population")
plt.title("Race distribution")
plt.xticks(sp.arange(6), ("Asian", "Black", "Hisp", "NatAm", "Other", "White"))
```

Question: What conclusion about the composition of the giant component can be drawn from this plot?

Answer: The giant component in many parts, holds a similar composition of the percentages from each race. It's indicative that the overall this social network's giant component mimics the characteristics of the entire network. In the giant component, there are less black races, and more white races in terms of percentages. This may indicate that since the giant component is composed of about the same percentages, that these groups are keeping the status quo amongs themselves. It's hard to determine segregation, but it's expected given that, if there is a lot less segregation, we could expect a more balanced percent of the population in the giant component rather than it being overwhelmingly white.

### Part e. Female/Male Degree Distribution (2 points)

Compute the degree distribution for the female and male students in the population. (Whole graph). (You can use `bincount` for this, too.)

```python
#Grab a dictionary of all the nodes with their gender tags
GenderN = nx.get_node_attributes(fmh, 'GenderN')
#Generate a degree list
fmhDeg = nx.degree(fmh)


#Grab all the degrees matching a female tag
fmhDegF = [item[1] for item in fmhDeg if GenderN.get(item[0]) == 0]
#repeat w/ all the males
fmhDegM = [item[1] for item in fmhDeg if GenderN.get(item[0]) == 1]

fBin = sp.bincount(fmhDegF)
mBin = sp.bincount(fmhDegM)
print('{} {}'.format("Females:", fBin))
print('{} {}'.format("  Males:", mBin))
```

```python
#Calculating fractions
fPercent = [(item/len(fmhDegF)) for item in fBin]
mPercent = [(item/len(fmhDegM)) for item in mBin]

```

Plot the degree distributions using a line plot. See Example Plot 2.

```python
plt.plot(fPercent, 'r.-', label = 'Female')
plt.plot(mPercent, 'b.-', label = 'Male')
plt.ylabel("% of group")
plt.xlabel("Degree")
plt.title("Degree distribution by gender")
plt.legend()
```

Question: What can you conclude from this plot?

Answer: There are a lot more males that have no connections in the social network than there are females. In fact, overall the females have a greater amount of degree, and there are some that have high degrees of seven or eight, demonstrating how well connected the females are incomparison to the males.

### Part f:	Segregation calculation (2 pts)


We will define the segregation of a node as the fraction of the neighbor nodes with the same Race attribute. Should return a dictionary with this value for each node. You may want to write an intermediate function that does the calculation for one node, but that function will have to take the graph as an argument because `neighbors()` is a Graph function. You will have to account for nodes with no neighbors. They should return an Inf value `float('inf')`.

```python
def nodeSegregation(Graph, node):
    #Graph is the full graph, node is name of node ie, n0,n1,n2...nk-1,nk
    count = total = 0
    if(len(list(Graph.neighbors(node)))== 0): return float('inf')
    for item in Graph.neighbors(node):
        #compare using the raceN dictionary if they are equal
        if(RaceN.get(item) == RaceN.get(node)):
            count+=1
        total+=1
    return count/total;
        
#quick test
nodeSegregation(fmh,'n30')
#fmh.node()
```

```python
def GraphSegregation(Graph):
    #nodesList = list(Graph.nodes())
    segDict = {node:nodeSegregation(Graph,node) for node in Graph.nodes()}
    return segDict
```

Using your function calculate the segregation values for the whole graph.

Calculate average segregation for male and female students. Ignore the singleton nodes. Use `sp.isinf()` to check for the Inf values. Hint: The value for male students should be between 0.80 and 0.81.

```python
NetworkSeg = GraphSegregation(fmh)
GenderDict = (nx.get_node_attributes(fmh, "Sex"))
totalF = segF = 0
totalM = segM = 0

for node in NetworkSeg.keys():
    #ignore inf nodes
    if(NetworkSeg.get(node) != float('inf')):
        sex = GenderDict.get(node)
        if(sex == 'F'):
            totalF += 1
            segF += NetworkSeg.get(node)
        else:
            #print(node)
            totalM +=1
            segM += NetworkSeg.get(node)
    
print(segF/totalF)
print(segM/totalM)
#print(NetworkSeg.get(node))
```
