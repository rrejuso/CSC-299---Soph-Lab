### Question 8
Random node removal

Write code to remove nodes randomly from each network, stopping when the largest component is less than 0.66 times the size (number of nodes) as the _original_ network. (Don't forget to make a copy of the network before you start removing nodes.) Return the number of removals required. Hint: the original network has 62 nodes, so you will stop when the giant component has 40 nodes or less (62 * 0.66 = 40.92). 2nd hint: I recommend that you decompose this problem into three or four distinct functions and work on one at a time. _2 points_

```python
#This code iterates over list of connected components and gets number of connected nodes
def checkLength(myGraph):
    largest_cc = max(nx.connected_components(myGraph), key=len)
    return len(largest_cc)

#Returns just the counter of random removal
def RandRemoveCounter(Graph):
    #track number of removals
    GraphCpy = Graph.copy()
    numRemovals = 0
    #minSize is 66% of the node size
    minSize = ( len(GraphCpy)*2) //3
    #Check while largest component is greater than minSize
    while(checkLength(GraphCpy) > minSize):
        #grab a random node and remove
        randomNode = random.choice(list(GraphCpy.nodes()))
        GraphCpy.remove_node(randomNode)
        numRemovals +=1
    return numRemovals

#Returns the graph of Random Removal
def RandRemoveGraph(Graph):
    GraphCpy = Graph.copy()
    #minSize is 66% of the node size
    minSize = ( len(GraphCpy)*2) //3
    #Check while largest component is greater than minSize
    while(checkLength(GraphCpy) > minSize):
        #grab a random node and remove
        randomNode = random.choice(list(GraphCpy.nodes()))
        GraphCpy.remove_node(randomNode)
    return GraphCpy
    

#DolphinRemoval = dolphin_multi.copy() 
print(f"Dolphin #removals: {RandRemoveCounter(dolphin_multi)} ")
print(f"ER #removals: {RandRemoveCounter(er_multi)}")
```

### Question 14
Targeted node removal

Write code to remove the largest degree nodes from each network, stopping when the largest component is 0.66 times the original network. As before, return the number of removals required.
_2 point_

```python
#Helper method for finding the largest Node
def GetLrgstDegree(Graph):
    #Graph
    max = 0
    for i in list(Graph.degree()):
        if i[1] > max:
            max = i[1]
            node = i[0]
    return node

#Returns the graph
def TrgtedNodeRemovalGraph(Graph):
    trgtCpy = Graph.copy()
    minSize = ( len(trgtCpy)*2) //3
    #Check while largest component is greater than minSize
    while(checkLength(trgtCpy) > minSize):
        #Remove the largest node
        trgtCpy.remove_node(GetLrgstDegree(trgtCpy))
    return trgtCpy

#Returns number of times a node was retmoved
def TrgtedNodeRemovalCounter(Graph):
    trgtCpy = Graph.copy()
    minSize = ( len(trgtCpy)*2) //3
    #Check while largest component is greater than minSize
    numRemovals = 0
    while(checkLength(trgtCpy) > minSize):
        #Remove the largest node
        trgtCpy.remove_node(GetLrgstDegree(trgtCpy))
        numRemovals+= 1;
    return numRemovals
```
