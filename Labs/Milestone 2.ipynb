```python

#Returns a tuple of (Graph, Color_map)
def getActorGraph(actor,endpoint,limit = -1):
    #insert the actor name for the dbr:{actor name} in the query
    limitString = ''  #By default, no limit
    if(limit == -1): limitString = ''''''
    else:
        limitString = '''LIMIT ''' + str(limit)
    
    query = ['''SELECT ?movie ?actor 
    WHERE { ?movie rdf:type dbo:Film .
    ?movie dbo:starring dbr:''',
             actor,''' . 
    ?movie dbo:starring ?actor}''',limitString,]
    query = ''.join(query) #combine the string
    myDataFrame = endpoint.select(query)
    #print(myDataFrame)
    myMovieList = set(myDataFrame.movie)
    #Generate graph as Directed Graph
    myGraph = nx.from_pandas_edgelist(myDataFrame, source ='actor',target='movie', edge_attr=True,create_using=nx.DiGraph())
    #Generate color list
    color_map = []
    for node in myGraph:
        if node in myMovieList:
            color_map.append('green')  #Movies are green
        else: color_map.append('red')  #Colors are blue
    return (myGraph, color_map)
    

    #To use this function, create a variable, put in the endpoint and actor name
    #VERY IMPORTANT, use the name found in the dbpedia website with underscores
    #for example, the site http://dbpedia.org/page/Dwayne_Johnson the person is Dwayne_Johnson
    #Same thing for other actors like http://dbpedia.org/page/Will_Ferrell   is Will_Ferrell
```

```python
#Important! The third argument is an optional argument that limits how much data we get
#There is no sorting, so the way they are returned is arbitrary
myData = getActorGraph("Dwayne_Johnson",endpoint)
nx.draw(myData[0], with_labels=True, node_color=myData[1])

#Export as a graphml! <BIG ERRORS HERE ON MINE, HOPEFULLY NOT ON YOURS>
#the 'tuple' is whatever you named your variable, in this case myData
#remember, the tuple[0] is the graph. tuple[1] is the color map, so you want
#so we want myData[0] to get the graph

#takes (Graph,<fileName>)

#<UNCOMMENT THIS LINE  VVV>

    #nx.write_graphml(myData[0], "OurGraph.graphml")

#<UNCOMMENT THAT LINE ^^^^
```
