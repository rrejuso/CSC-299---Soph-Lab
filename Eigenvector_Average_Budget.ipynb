```python
import networkx as nx
from math import sqrt
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline 
```

```python
dwayne_budget_file = "AverageBudgetDw.graphml"
```

```python
dwayne = nx.read_graphml(dwayne_budget_file)
```

```python
names = {node : dwayne.node[node]['label']for node in dwayne.nodes()}
nx.draw(dwayne, with_labels=True, labels=names)
```

```python
dwayne_eig = dict(nx.eigenvector_centrality(dwayne, weight="weight"))
print(dwayne_eig)
```

```python
plt.title ("Eigenvector / Dwanye Budget")
bins = [0, 0.000000001, 0.000000002 , 
        0.000000003, 0.000000004, 
        0.000000005,0.000000006, 0.000000007,
        0.000000008, 0.000000009, 0.000000010, 
        0.000000011, 0.000000012, 0.000000013]
plt.hist(dwayne_eig.values(), bins, rwidth = 0.8)
```
