```python
AverageBudget= []
for iterator in neighbors:
    total = 1;
    sum = 0;
    for item in iterator:
        if math.isnan(dwayneDict[item]['budget']) == False:
            total= total + 1
            sum = sum + dwayneDict[item]['budget']
            #print("{},{}".format(item, dwayneDict[item]['budget']))
    avg = sum/total
    #print(avg)
    AverageBudget.append(avg)

    
AverageBudget
```

```python
AverageGross= []
neighbors = [dwayne.neighbors(node) for node,data in dwayne.nodes(data=True) if data['bipartite'] == 0]
for iterator in neighbors:
    total = 1;
    sum = 0;
    for item in iterator:
        if math.isnan(dwayneDict[item]['gross']) == False:
            total= total + 1
            sum = sum + dwayneDict[item]['gross']
            #print("{},{}".format(item, dwayneDict[item]['budget']))
    avg = sum/total
    #print(avg)
    AverageGross.append(avg)

AverageGross
```

```python
#len(AverageBudget) == len(actors)
names = {actor:dwayneDict[actor]['label'] for actor in actors}
#for actor in actors:
#    names.append(dwayneDict[actor]['label'])
abDict = dict(zip(actors, AverageBudget))
agDict = dict(zip(actors,AverageGross))
#len(abDict)
#len(names)
```

```python
 #bipartite creation and filtering from lecture
actors = [node for node, node_data in dwayne.nodes(data=True) if node_data['bipartite']==0]
actorProjection = nx.bipartite.weighted_projected_graph(dwayne, actors)

filtered = nx.Graph(actorProjection)
```

```python
nx.set_node_attributes(filtered, abDict, "AverageBudget")
nx.set_node_attributes(filtered, names, "names")
nx.set_node_attributes(filtered, agDict, "AverageGross")
```

```python
AverageBudget = nx.get_node_attributes(filtered, "AverageBudget")

items = [AverageBudget[item] for item in AverageBudget]  
itemsFilter = [item for item in items if item != 0]
plt.hist(itemsFilter)
plt.yscale('log')
    
```

```python
AverageGross = nx.get_node_attributes(filtered, "AverageGross")

items = [AverageGross[item] for item in AverageGross]  
items
#plt.hist(itemsFilter)
#plt.yscale('log')
```

```python
to_remove = [(n1, n2) for n1, n2, data in actorProjection.edges(data=True) if data['weight'] < 2]

filtered.remove_edges_from(to_remove)

giant_cc_nodes = max(nx.connected_components(filtered), key=len)
filtered2 = filtered.subgraph(giant_cc_nodes)
```

```python
dwayne_deg = dict(nx.degree(filtered2, weight="weight"))
dwayne_bet = dict(nx.betweenness_centrality(filtered2, weight="weight"))
dwayne_clo = dict(nx.closeness_centrality(filtered2))
```

```python
dwayne_df = pd.DataFrame({"Wt Degree": dwayne_deg, "Betweenness":dwayne_bet, "Closeness":dwayne_clo, "Names": names})
dwayne_df = dwayne_df.set_index("Names")
dwayne_df.head()
```
